```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Visulizing data

```python
data = pd.read_csv("../data/animals.csv",header=0)
```

```python
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="height", y="weight", hue="walks_on_n_legs", palette="Set2")

plt.title("Weigh vs height")
plt.xlabel("height (m)")
plt.ylabel("Weigh (kg)")
plt.legend(title="walks_on_n_legs")
plt.grid(True)
plt.show()
```

```python
count = data[(data["has_wings"] == False) & (data["has_tail"] == False)].shape[0]

print(f"Number of animals without wings and Tail: {count}")
```

```python
count = data[(data["has_wings"] == True) & (data["has_tail"] == False)].shape[0]

print(f"Number of animals wit wings but without tail: {count}")
```

```python
count = data[data["walks_on_n_legs"].isin([1, 3, 5])].shape[0]

print(count)


count_not_in = data[~data["walks_on_n_legs"].isin([1, 3, 5])].shape[0]
print(f"With 2 or 4 legs: {count_not_in}")
```

```python
count = data[data["walks_on_n_legs"].isin([1, 3, 5])].shape[0]

print(count)
```

```python
# Filtrar los datos con peso menor a 2000 kg
filtered_data = data[data["weight"] < 2000]

# Graficar
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_data, x="height", y="weight", hue="walks_on_n_legs", palette="Set2")

plt.title("Weight vs Height (Filtered by Weight < 2000 kg)")
plt.xlabel("Height (m)")
plt.ylabel("Weight (kg)")
plt.legend(title="Walks on N Legs")
plt.grid(True)
plt.show()
```

#### **Conclusion**: Consider the outlimers values are that animals: 
#### - Have 1, 3, or 5 legs
#### - Have neither wings nor a tail
#### - Have a height greater than 4.5 meters
#### - Have wings but no tail (these are assumed to be other types of animals)

### Working with DataLabeled

```python
dataClean = pd.read_csv("../data/animalsLabeled.csv",header=0)
```

```python
import matplotlib.patches as mpatches

plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataClean, x="height", y="weight", hue="label", palette="Set2")

palette = sns.color_palette("Set2")

handle1 = mpatches.Patch(color=palette[0], label="1 - Kangaroo")
handle2 = mpatches.Patch(color=palette[1], label="2 - Elephant")
handle3 = mpatches.Patch(color=palette[2], label="3 - Chicken")
handle4 = mpatches.Patch(color=palette[3], label="4 - Dog")



plt.title("Weigh vs height")
plt.xlabel("height (m)")
plt.ylabel("Weigh (kg)")
plt.legend(handles=[handle1, handle2, handle3, handle4], title="Label")
plt.grid(True)
plt.show()

```
